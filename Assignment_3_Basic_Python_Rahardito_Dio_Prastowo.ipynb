**5. Magnitude Scale**

Perhatikan data di bawah ini terkait magnitude gempa dan deskripsinya:

Buatlah program yang membaca input magnitude gempa dari user dan program menunjukkan informasi deskripsi dari gempa tersebut. Contoh: Jika user memberi input 5.5 maka output dari program adalah “Moderate earthquake


```python
def get_magnitude(n):
    if 2.0 <= n < 3.0:
        return 'Minor'
    elif 3.0 <= n < 4.0:
        return 'Minor'
    elif 4.0 <= n < 5.0:
        return 'Light'
    elif 5.0 <= n < 6.0:
        return 'Moderate'
    elif 6.0 <= n < 7.0:
        return 'Strong'
    elif 7.0 <= n < 8.0:
        return 'Major'
    elif 8.0 <= n < 9.0:
        return 'Great'
    elif 9.0 <= n < 10.0:
        return 'Great'
    elif 10.0 <= n:
        return 'Great'
    else:
        return 'Insert more than 2.0 Magnitude'

Scale = float(input("Insert Scale: "))
print(get_magnitude(Scale) + " Earthquake")
```

2. Buatlah program untuk menghitung regresi linier menggunakan metode invers matriks
x = (A^T * A)^(-1) * A^T * y
dimana :
- x adalah koefisien dari model regresi yang ingin dicari
- A adalah matriks data
- A^T adalah transpose dari A
- A^(-1) adalah invers dari matriks A
- y adalah variabel target

```python
# get colomn values
x = dataset.X.values
y = dataset.Y.values

# create matrix invers
x_matrix = np.vstack((np.ones(len(x)), x)).T
x_invers = np.linalg.inv(x_matrix.T.dot(x_matrix)).dot(x_matrix.T).dot(y)

# get output predictions 
y_pred = x_matrix.dot(x_invers)

# plot data and predictions
plt.scatter(x, y,label="data");
plt.plot(x, y_pred, color='red',label="fitted line");

# create blue text at x=0.25 and y=3
plt.text(0.25,3, r"$\beta_0 = 3.99187633$", fontsize=14, color="b")
# create blue text at x=0.25 and y=3
plt.text(0.25,2, r"$\beta_1 = 2.98060191$", fontsize=14, color="b")
plt.legend()
plt.show()
```

3. Buatlah program untuk menghitung regresi linier menggunakan **method linalg.lstsq** dari numpys

```python
# get variable colomn
x = dataset.X
y = dataset.Y

# linearly generated sequence
A = np.array([x, np.ones(x.shape[0])])

# obtaining the parameters of regression line
w = np.linalg.lstsq(A.T, y, rcond=None)[0]

# regression line
line = w[0]*x + w[1]  
plt.plot(x, line, 'r-',label="fitted line")
plt.plot(x, y, 'o',label='data')

# create blue text at x=0.25 and y=3
plt.text(0.25,3, r"$\beta_0 = 3.99187633$", fontsize=14, color="b")
# create blue text at x=0.25 and y=3
plt.text(0.25,2, r"$\beta_1 = 2.98060191$", fontsize=14, color="b")
plt.legend()
plt.show()
```
